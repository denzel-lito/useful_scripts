# определение кол-ва оптимального кол-ва кластеров

```python
inertia = []
for k in tqdm(range(1,16)):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(df_tab.iloc[:,1:])
    inertia.append(np.sqrt(kmeans.inertia_))

plt.figure(figsize=(20,12))
plt.plot(range(1,16), inertia, marker='s')
plt.xticks(range(1,16))
plt.xlabel('k')
plt.ylabel('J(C_k)')
```

```python
for i in range(1,len(inertia)):
    print(inertia[i]-inertia[i-1])
```

```python
model = KMeans(n_clusters=3, random_state=1, n_init=50, max_iter=10000, verbose=0).fit(df_tab)
df_tab['cluster'] = model.predict(df_tab)
df_tab.sort_values('cluster', inplace=True)
pd.value_counts(df_tab['cluster']).sort_index()
```
